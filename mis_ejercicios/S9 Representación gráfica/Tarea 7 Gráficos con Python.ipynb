## Ejercicio 1

A partir del siguiente c칩digo, intenta obtener el gr치fico del dibujo llamado ejercicio 1 en el enunciado.

Pista: investiga el par치metro **fill_between** de matplotlib.

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X) 

plt.fill_between(X, Y + 1, 1, color='b', alpha=0.2)

plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.fill_between(X, Y - 1, -1, where=(Y - 1 > -1), color='b', alpha=0.2)
plt.fill_between(X, Y - 1, -1, where=(Y - 1 < -1), color='r', alpha=0.2)

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.axis([-np.pi, np.pi, -2.5, 2.5])

plt.plot(X, Y - 1, color='blue', alpha=1.00)
```

## Ejercicio 2

A partir del siguiente c칩digo, intenta obtener el gr치fico del dibujo llamado ejercicio 2 en el enunciado.

Pista: el color viene dado por el 치ngulo que forman x e y con la horizontal

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

import math
angulo = np.array([math.atan(z) for z in Y/X])

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.axis([-1, 1, -1, 1])

plt.scatter(X,Y, c=angulo)
```

## Ejercicio 3

A partir del siguiente c칩digo, intenta obtener el gr치fico del dibujo llamado ejercicio 3 en el enunciado.

Pista: cuidado con la alineaci칩n del texto en el centro de las barras.

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')  

ax = plt.gca()
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.axis([-1, 12, -1.2, 1.2])

for i, j in zip(X, Y1):
    plt.annotate(round(j, 2), 
                 xy=(i, j), 
                 xytext=(0, 3), 
                 textcoords="offset points", 
                 ha='center', 
                 va='bottom'
                )
    
for i, j in zip(X, Y2):
    plt.annotate(round(j, 2), 
                 xy=(i, -j), 
                 xytext=(0, -3), 
                 textcoords="offset points", 
                 ha='center', 
                 va='top'
                )
```

## Ejercicio 4

A partir del siguiente c칩digo, intenta obtener el gr치fico del dibujo llamado ejercicio 4 en el enunciado.

Pista: tendr치s que modificar la Z.

```python
Z = np.random.uniform(0, 1, 20)

explode = [0.1] * len(Z)
explode[len(Z)-1] = 0.2
colors = plt.cm.gray(np.linspace(0,1,20))

plt.pie(
    Z, 
    explode=explode, 
    colors=colors, 
    wedgeprops = dict(
        linewidth = 1, 
        edgecolor='k'), 
    radius=2
)

plt.show()
```

```python
fig = plt.figure(111,figsize = (12,8))

Z = np.ones(20)
Z[19] = 2
explode = np.ones(20)*0.05
explode[19] = 0.1
colors = np.linspace(0,255,20)

plt.pie(Z, explode = explode,colors = [str(x/255) for x in colors],wedgeprops = {'edgecolor': 'black'})

plt.xticks([])
plt.yticks([])
plt.show()
```

```python
fig, ax = plt.subplots()

size = 0.3
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Pie plot with `ax.pie`')
plt.show()
```
