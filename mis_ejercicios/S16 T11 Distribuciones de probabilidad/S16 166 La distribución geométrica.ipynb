# Distribución geométrica

```python
from scipy.stats import geom
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
p = 0.3  # 30 % de posibilidades de abrir la puerta (3 llaves de 10 abren la puerta).

mean, var, skew, kurt = geom.stats(p, moments='mvsk')

print('Media    : %f'%mean)
print('Varianza : %f'%var)
print('Sesgo    : %f'%skew)
print('Curtosis : %f'%kurt)

x = np.arange(geom.ppf(0.01, p), geom.ppf(0.99, p))
ax.plot(x, geom.pmf(x, p), 'bo', ms=8, label='Distribución Geom(0.3)')
ax.vlines(x, 0, geom.pmf(x, p), colors='b', lw=4, alpha=0.5)

rv = geom(p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='--', lw=1, label='Frozen PMF')
ax.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots()
prob = geom.cdf(x, p)

ax.plot(x, prob, 'bo', ms=8, label='Función de distribución acumulada')
plt.show()
```

```python
fig, ax = plt.subplots()
r = geom.rvs(p, size=10000)
plt.hist(r)
plt.show()
```
