# Distribución binomial

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1)
n = 7
p = 0.4
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')

print('Media    : %f'%mean)
print('Varianza : %f'%var)
print('Sesgo    : %f'%skew)
print('Curtosis : %f'%kurt)

x = np.arange(binom.ppf(0.0001, n, p), binom.ppf(0.999, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='Función de densidad de B(7, 0.4)')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=4, alpha=0.5)
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='--', lw=1, label='Distribución teórica')
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
n = 7
p = 0.4
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')

print('Media    : %f'%mean)
print('Varianza : %f'%var)
print('Sesgo    : %f'%skew)
print('Curtosis : %f'%kurt)

x = np.arange(binom.ppf(0.0001, n, p), binom.ppf(0.999, n, p))
ax.plot(x, binom.cdf(x, n, p), 'bo', ms=8, label='Función de densidad acumulada de B(7, 0.4)')
ax.vlines(x, 0, binom.cdf(x, n, p), colors='b', lw=4, alpha=0.5)
rv = binom(n, p)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='--', lw=1, label='Distribución teórica acumulada')
ax.legend(loc='best', frameon=False)
plt.show()
```
