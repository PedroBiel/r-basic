# Gráficos desde Data Frames y exportación

```python
import pandas as pd
import numpy as np
```

```python
years = [year for year in range(1900, 2020)]
deads = [(y + np.random.uniform(0, 100) - 1850) for y in years]
```

```python
df = pd.DataFrame({
    'year': years,
    'deads': deads
})
```

```python
df.head()
```

```python
df.plot(kind='scatter', x='year', y='deads', figsize=(10, 8))
```

```python
my_plot = df.plot(kind='scatter', x='year', y='deads', figsize=(10, 8))
my_fig = my_plot.get_figure()
```

```python
my_fig.savefig('../S10 T4 Data frames/dataframe_plot.png')
```
