# Histogramas en Python

## Histogramsa con Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
size, sclae = 1000, 10
data = pd.Series(np.random.gamma(sclae, size=size))
```

```python
sns.histplot(data, kde=True)
```

```python
np.random.seed(2019)
x = np.random.laplace(loc=10, scale=3, size=1000)
sns.set_style('darkgrid')
sns.histplot(x, kde=True)
```

```python
from scipy import stats
```

```python
sns.distplot(data, fit=stats.laplace)
```

## Otras formas

```python
data2 = np.random.choice(np.arange(10), size=10000, p=np.linspace(1, 11, 10) / 60)
```

```python
s = pd.Series(data2)
```

```python
s.value_counts()
```

```python
s.value_counts(normalize=True)
```

```python
ages = pd.Series([1, 1, 3, 5, 6, 8, 9, 10, 12, 15, 18 ,18 ,18, 20, 25, 30, 40, 51, 52])
bins = (0, 10, 15, 18, 21, np.inf)
labels = ('infancia', 'preadolescecnia', 'adolescencia', 'universitario', 'adulto')
groups = pd.cut(ages, bins=bins, labels=labels)
groups
```

```python
groups.value_counts()
```

```python
pd.concat((ages, groups), axis=1).rename(columns={0:'age', 1:'group'})
```
